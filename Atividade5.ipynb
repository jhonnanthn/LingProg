# 1 Linha: Crie a classe Linha que tem dois atributos, coordenada1 e coordenada2. Cada coordenada é uma tupla que carrega duas coordenadas cartesianas (x,y) que denotam pontos do segmento de reta. Faça métodos que calculem o comprimento do segmento de reta e sua inclinação.

```python
class Linha:
    
    def __init__(self, coordenadas1, coordenadas2):
        self.coordenadas1 = coordenadas1
        self.coordenadas2 = coordenadas2
        
    def comprimento(self, coordenadax, coordenaday):
        coordenadax * coordenaday
```

```python
linha = Linha(12,12)
linha.comprimento(12,12)
```

# 2 Figuras: Crie a seguinte hierarquia de classes de figuras geométricas. Veja na figura as fórmulas:
a. A classe abstrata Figura deve ter o método abstrato area.
b. A classe concreta Circulo é subclasse de Figura.
c. A classe abstrata Poligono é subclasse de Figura e deve ter os atributos base e
altura .
d. As classes concretas Triangulo, Losango, Retangulo e Quadrado são subclasses
de Poligono. Tente criar mais uma generalização aqui olhando as fórmulas da área.
e. Os polígonos Retangulo e Quadrado devem implementar a interface Diagonal,
que deve ter um método que calcula a diagonal.
f. Crie uma classe Geometria com uma lista de Figuras com pelo menos uma
figura de cada e imprima suas áreas, perímetros e diagonais.

```python
class Figura:
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura
    
    def area(self, base, altura):
        return base*altura
        
class Circulo(Figura):
    def area(self, base, altura):
        return base*altura

class Poligono(Figura):
    def __init__(self, name, base, altura):
        self.name = name
        self.base = base
        self.altura = altura
    def area(self, base, altura):
        return base*altura
    
class Triangulo(Poligono):
    def area(self, base, altura):
        return base*altura
class Losangulo(Poligono):
    def area(self, base, altura):
        return base*altura
class Retangulo(Poligono):
    def area(self, base, altura):
        return base*altura
class Quadrado(Poligono):
    def area(self, base, altura):
        return base*altura
```

```python
figura = Figura(3,2)
print(figura.area)
```
